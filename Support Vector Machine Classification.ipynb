```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import the dataset
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
#Splitting the data into indepandant and depandant variables
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
#Spltting the data into test and training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Training the SVM model
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear') #Must specify weather want a linear or non-linear kernel
classifier.fit(X_train,y_train)
```

> The **C parameter** tells the SVM optimization how much you want to avoid misclassifying each training example. For large values of C, the optimization will choose a smaller-margin hyperplane if that hyperplane does a better job of getting all the training points classified correctly. Conversely, a very small value of C will cause the optimizer to look for a larger-margin separating hyperplane, even if that hyperplane misclassifies more points. For very tiny values of C, you should get misclassified examples, often even if your training data is linearly separable.

```python
#Predict results
classifier.predict(sc.transform([[20,20000]]))
```

```python
y_pred = classifier.predict(X_test)
y_pred = y_pred.reshape(len(y_pred),1)
y_pred
```

```python
y_test = y_pred.reshape(len(y_test),1)
y_test
```

```python
np.concatenate((y_pred,y_test),1)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
#Visualise the train set
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
